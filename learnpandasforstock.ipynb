第一步，先獲得要分析的數據：
以下代碼可以從yahoo獲取股票相關的數據

輸入分析時段
    開始時間 startdate
    結束時間 enddate

```python
startdate='2012/06/30'
enddate='2013/12/30'
```

```python
import pandas.io.data as web
from pandas import DataFrame
data_feed = {}
symbols=['AAPL','FB', 'SPLK', 'YELP', 'GG','BP','0050.TW','^TWII', '5434.TW','2330.TW']
for ticker in symbols:
    data_feed[ticker] = web.get_data_yahoo(ticker, startdate, enddate)
    
price = DataFrame({tic: data['Adj Close'] for tic, data in data_feed.iteritems()})
Volume = DataFrame({tic: data['Volume'] for tic, data in data_feed.iteritems()})
returns = price.pct_change()
```

#第二步：簡單的收益分析。
我們可以用dataframe的sum函數來對收益率進行匯總，並以條形圖的形式來對比分析多隻股票的收益率。

##圖1.年收益率結果

```python
import matplotlib.pyplot as plt
returns.sum().plot(kind='bar',title="% return For Year")
plt.show()
```

如圖1 所示，Facebook 年初至今它的損失接近40%。相比之下，Yelp收益接近40%。這樣看來，賣空Facebook 並買進Yelp是明智的選擇。

上面這個圖只是演示下，實際上我們很難從這個圖中看出什麼結果。不過我們卻可以看下累計的收益率走勢圖，看下是否有明顯的特徵。計算累計收益率也非常簡單，直接調用cumsum 函數就可以了。
代碼如下：
ts = returns.cumsum()

###圖4. 累計收益率

```python
ts = returns.cumsum()
ts.plot()
plt.legend(loc='upper left')
plt.show()
```

圖4 顯示了更多的信息，比如Facebook ，在9月份的時候累計收益率曾經下降到了60%。而且Facebook 的收益率波動也比較大，這也反映出了比較高的風險水平。所以，如果你想持股Facebook 的話，一定要控制好投資比重。
 
當然，我們也可以直接查看下多隻股票收益率之間的相關性，直接用corr函數。

```python
market_data_feed = {}
market_symbols=['SPY']
for ticker in market_symbols:
    market_data_feed[ticker] = web.get_data_yahoo (ticker, '5/21/2012', '11/1/2012')
market_price = DataFrame({tic: data['Adj Close'] for tic, data in market_data_feed.iteritems()})
market_volume = DataFrame({tic: data['Volume'] for tic, data in market_data_feed.iteritems()})
market_returns = market_price.pct_change()
mts = market_returns.cumsum()
mts.plot(); plt.legend(loc='upper left')
plt.show()
```
